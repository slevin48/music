## Manipulate bucket as a file

https://s3fs.readthedocs.io/

```python
import s3fs
fs = s3fs.S3FileSystem(anon=False)
```

```python
fs.ls('music48')
```

```python
m = fs.ls("music48/Miley Cyrus/Bangerz [Deluxe Edition]")
m[1]
```

```python
with fs.open(m[1],"rb") as f:
    n = f.read()
```

```python
with open("downloads/"+m[1].replace("music48/Miley Cyrus/Bangerz [Deluxe Edition]",""),"wb") as f:
    f.write(n)
```

## Store Music in S3 bucket

```python
import boto3
title = video.title
s3_client = boto3.client('s3')
file_name = "downloads/"+title+".mp3"
object_name = file_name
s3_bucket = "music48"
# s3_client.upload_file(file_name, bucket, object_name)
s3_client.upload_file(file_name, s3_bucket, object_name)
```

```python
import pytube, os, boto3

def up_music(search):
    # search = "someday one republic"
    r = pytube.Search(search).results
    video = r[0].streams.filter(only_audio=True).first()
    title = video.title
    out_file = video.download('downloads')
    # Video to Audio
    base, ext = os.path.splitext(out_file)
    new_file = base + '.mp3'
    if not os.path.exists(new_file):
        os.rename(out_file, new_file)
    # Upload to S3
    s3_client = boto3.client('s3')
    file_name = "downloads/"+title+".mp3"
    object_name = file_name
    s3_bucket = "music48"
    s3_client.upload_file(file_name, s3_bucket, object_name)
```

```python
search = "adele easy on me"
up_music(search)
```
